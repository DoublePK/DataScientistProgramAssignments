## Import Required Libraries

```python
from sklearn.metrics import mean_squared_error as MSE
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.tree import DecisionTreeRegressor
```

### Read How many Hours Play the game based on Weather Condition

```python
data = pd.read_csv('hours_played.csv')
data
```

### Change Categorical variable to Numeric Format

```python
data['Outlook'].replace(['Sunny', 'Overcast', 'Rainy'],[1,2, 3], inplace = True)
data['Temp'].replace(['Hot','Mild','Cool'],[1, 2, 3], inplace = True)
data['Humidity'].replace(['High', 'Normal'], [1, 2], inplace = True)
data['Windy'].replace(['True', 'False'], [1, 2], inplace = True)
```

### Separate Feature and target   data

```python
X = data[['Outlook', 'Temp', 'Humidity', 'Windy']]
y = data['Hours Played']
```

## Fit a Decision  Tree for Regressiion with MSE impurity

### Visualize the MSE model

```python
from sklearn import tree

plt.figure(figsize = (15,10))
_ = tree.plot_tree(dt, 
                   feature_names=['Outlook', 'Temp', 'Humidity', 'Windy'],  
                   class_names= ['Hours Played'],
                   filled=True)
```

## Fit a Decision  Tree for Regressiion with MAE impurity

```python
dt = DecisionTreeRegressor(criterion = 'absolute_error',min_samples_leaf = 0.14,random_state=3)

dt.fit(X, y)

y_prediction = dt.predict(X)

print("Train MSE: {:.2f}".format(MSE(y, y_prediction)))
```

### Visualize the MAE model

```python
plt.figure(figsize = (15,10))
_ = tree.plot_tree(dt, 
                   feature_names=['Outlook', 'Temp', 'Humidity', 'Windy'],  
                   class_names= ['Hours Played'],
                   filled=True)
```

## Fit a Decision  Tree for Regressiion with Friedman MSE impurity

```python
dt = DecisionTreeRegressor(criterion = 'friedman_mse',min_samples_leaf = 0.14,random_state=3)

dt.fit(X, y)

y_prediction = dt.predict(X)

print("Train MSE: {:.2f}".format(MSE(y, y_prediction)))
```

### Visualize the Friedman MSE model

```python
plt.figure(figsize = (15,10))
_ = tree.plot_tree(dt, 
                   feature_names=['Outlook', 'Temp', 'Humidity', 'Windy'],  
                   class_names= ['Hours Played'],
                   filled=True)
```

## Fit a Decision  Tree for Regressiion with Poisson impurity

```python
dt = DecisionTreeRegressor(criterion = 'poisson',min_samples_leaf = 0.14,random_state=3)

dt.fit(X, y)

y_prediction = dt.predict(X)

print("Train MSE: {:.2f}".format(MSE(y, y_prediction)))
```
