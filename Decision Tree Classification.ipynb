## Import Required Libraries

```python
from sklearn.model_selection import  train_test_split 
from sklearn.metrics import  accuracy_score
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

### Import Decision Tree Classifier

```python
from sklearn.tree  import  DecisionTreeClassifier 
```

### Read Golf Play DataSet in CSV format

```python
data = pd.read_csv('datasets/play_golf.csv')
data
```

### Transform Caterigorical variable to Numeric froamt

```python
data['Outlook'].replace(['Sunny', 'Overcast', 'Rain'],[0,1,2], inplace = True)
data['Temperature'].replace(['Hot','Mild','Cool'],[0, 1, 2], inplace = True)
data['Humidity'].replace(['High', 'Normal'], [0,1], inplace = True)
data['Wind'].replace(['Strong', 'Weak'], [0,1], inplace = True)
data['Play Golf'].replace(['No', 'Yes'], [0, 1], inplace = True)
```

### Separate Data to Features and Target Data

```python
X = data[['Outlook', 'Temperature', 'Humidity', 'Wind']]
y = data['Play Golf']
```

## Train Data with 'Gini' criterion with maximum depth of 2

```python
gini_model = DecisionTreeClassifier(criterion = 'gini', max_depth =  2, random_state = 1)
gini_model.fit(X, y)

gini_pred = gini_model.predict(X)
gini_acc = accuracy_score(y, gini_pred)
print('Accuracy of model with Gini impurity: {:.3f}'.format(gini_acc))
```

## Plot Tree Model with Gini Impurity

```python
from sklearn import tree

plt.figure(figsize = (15,10))
_ = tree.plot_tree(gini_model, 
                   feature_names=['Outlook', 'Temperature', 'Humidity', 'Wind'],  
                   class_names= ['No','Yes'],
                   filled=True)
```

## Train Data with 'Entropy' criterion with maximum depth of 2

```python
entropy_model = DecisionTreeClassifier(criterion = 'entropy', max_depth =  2, random_state = 1)
entropy_model.fit(X, y)

entropy_pred = entropy_model.predict(X)
entropy_acc = accuracy_score(y, entropy_pred)
print('Accuracy of model with Entropy impurity: {:.3f}'.format(entropy_acc))
```

## Plot Tree Model with Entropy Impurity

```python
plt.figure(figsize = (15,10))
_ = tree.plot_tree(entropy_model, 
                   feature_names=['Outlook', 'Temperature', 'Humidity', 'Wind'],  
                   class_names= ['No','Yes'],
                   filled=True)
```
